```python
import os
import re
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
AirData_16to19_Df = pd.read_csv(r"G:\Uppsala Docs\period5&6\ProcessedData\2016-2019_sensorData.csv")
```

```python
# drop unwanted column
AirData_16to19_Df = AirData_16to19_Df.drop(['Unnamed: 0','Start','National Station Code','Longitude','Latitude'],1)
# Rename Stop to Time
AirData_16to19_Df = AirData_16to19_Df.rename(columns={"Stop": "Timestamp"})
# convert columns to date
AirData_16to19_Df['Timestamp'] = pd.to_datetime(AirData_16to19_Df['Timestamp'])
AirData_16to19_Df.head(3)
```

```python
AirData_16to19_Df.describe()
```

```python
AirData_16to19_Df.info()
```

```python
# percentage of columns that are not empty
total = AirData_16to19_Df.isnull().sum().sort_values(ascending = False)
total = total[AirData_16to19_Df.isnull().sum().sort_values(ascending = False) != 0]
percent =  100 - total / len(AirData_16to19_Df) * 100
percent[AirData_16to19_Df.isnull().sum().sort_values(ascending = False) != 0].plot.bar(figsize = (20,10))
```

```python
# correlation Matrix with Heatmap
X = AirData_16to19_Df.iloc[:,9]
y = AirData_16to19_Df.iloc[:,-1]
# get correlations of each features in dataset
corrmat = AirData_16to19_Df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(10,10))
# plot heat map
g=sns.heatmap(AirData_16to19_Df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
# updating columns based on heatmap
AirData_16to19_Df = AirData_16to19_Df[['Timestamp', 'Classification', 'Station Name', 'NO2', 'NOX as NO2', 'PM2.5', 'Black Carbon','PM10']]
AirData_16to19_Df
```

```python
# copy air data
AirData_Copy = AirData_16to19_Df.copy()
```

```python
# choosing stations
AirData_Copy['Year'] = AirData_Copy['Timestamp'].dt.year
AirData_Copy['Month'] = AirData_Copy['Timestamp'].dt.month
# drop 2020 data
AirData_Copy = AirData_Copy[AirData_Copy['Year'] !=  2020]
```
