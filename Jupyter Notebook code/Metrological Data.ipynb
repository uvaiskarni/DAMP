```python
import os
import re
import pandas as pd
import numpy as np
```

```python
dataPath = r'G:\Uppsala Docs\period5&6\Meteorological data'
FeaturesPath = os.listdir(dataPath)
print(FeaturesPath)
```

## Air Pressure

```python
# reading air pressure data
filePath = dataPath+"\\"+FeaturesPath[0]
AirPressure = pd.read_csv(filePath, sep=";", skiprows=8, parse_dates=[['Datum', 'Tid (UTC)']],
                             usecols = ['Datum', 'Tid (UTC)','Lufttryck reducerat havsytans nivå','Kvalitet'])
print(AirPressure.describe())
print(AirPressure.info())
print(AirPressure.head(3))
```

```python
# renaming the columns
AirPressure.rename(columns={"Datum_Tid (UTC)":"DateTimestamp"}, inplace = True)
AirPressure.rename(columns = lambda x: re.sub('Lufttryck reducerat havsytans nivå','Air pressure',x), inplace = True)
AirPressure.rename(columns = lambda x: re.sub('Kvalitet','Quality',x), inplace = True)
# filter required data
AirPressure = AirPressure[AirPressure['DateTimestamp'] >= "2015-01-01 00:00:00"] 
print(AirPressure.describe())
print(AirPressure.info())
print(AirPressure.head(3))
print(AirPressure.groupby(['Quality']).count())
AirPressure = AirPressure[['DateTimestamp','Air pressure']]
```

## Amount of Clouds

## Preciptation Amount

```python
# reading air pressure data
filePath = dataPath+"\\"+FeaturesPath[3]
PreciptationAmount = pd.read_csv(filePath, sep=";", skiprows=8,
                                 usecols = ['Till Datum Tid (UTC)', 'Representativt dygn','Nederbördsmängd','Kvalitet'])

# convert columns to date
PreciptationAmount['Till Datum Tid (UTC)'] = pd.to_datetime(PreciptationAmount['Till Datum Tid (UTC)'])

print(PreciptationAmount.describe())
print(PreciptationAmount.info())
print(PreciptationAmount.head(3))
```

```python
# renaming the columns
PreciptationAmount.rename(columns={"Till Datum Tid (UTC)":"DateTimestamp"}, inplace = True)
PreciptationAmount.rename(columns = lambda x: re.sub('Representativt dygn','Representative day',x), inplace = True)
PreciptationAmount.rename(columns = lambda x: re.sub('Nederbördsmängd','Precipitation amount',x), inplace = True)
PreciptationAmount.rename(columns = lambda x: re.sub('Kvalitet','Quality',x), inplace = True)
# filter required data
PreciptationAmount = PreciptationAmount[PreciptationAmount['DateTimestamp'] >= "2015-01-01 00:00:00"] 
print(PreciptationAmount.describe())
print(PreciptationAmount.info())
print(PreciptationAmount.head(3))
print(PreciptationAmount.groupby(['Quality']).count())
PreciptationAmount = PreciptationAmount[['DateTimestamp','Precipitation amount']]
```

## Wind Direction

```python
# reading air pressure data
filePath = dataPath+"\\"+FeaturesPath[4]
# PreciptationAmount = 
WindDirection = pd.read_csv(filePath, sep=";", skiprows=9, parse_dates=[['Datum', 'Tid (UTC)']],
                  usecols = ['Datum','Tid (UTC)','Vindriktning','Kvalitet','Vindhastighet','Kvalitet.1'])

print(WindDirection.describe())
print(WindDirection.info())
print(WindDirection.head(3))
```

```python
# renaming the columns
WindDirection.rename(columns={"Datum_Tid (UTC)":"DateTimestamp"}, inplace = True)
WindDirection.rename(columns = lambda x: re.sub('Vindriktning','Wind direction',x), inplace = True)
WindDirection.rename(columns = lambda x: re.sub('Kvalitet','Quality',x), inplace = True)
WindDirection.rename(columns = lambda x: re.sub('Vindhastighet','Wind speed',x), inplace = True)
WindDirection.rename(columns = lambda x: re.sub('Kvalitet.1','Quality1',x), inplace = True)
# filter required data
WindDirection = WindDirection[WindDirection['DateTimestamp'] >= "2015-01-01 00:00:00"] 
print(WindDirection.describe())
print(WindDirection.info())
print(WindDirection.head(3))
```

```python
directions = np.array('N NNE NE ENE E ESE SE SSE S SSW SW WSW W WNW NW NNW N'.split())
bins = np.arange(11.25, 372, 22.5)
WindDirection['Wind direction'] = directions[np.digitize(WindDirection['Wind direction'], bins)]
# filter required data
WindDirection = WindDirection[WindDirection['DateTimestamp'] >= "2015-01-01 00:00:00"] 
print(WindDirection.describe())
print(WindDirection.info())
print(WindDirection.head(3))
print(WindDirection.groupby(['Quality']).count())
print(WindDirection.groupby(['Quality.1']).count())
WindDirection = WindDirection[['DateTimestamp','Wind direction','Wind speed']]
```

## Merging Additional data 

```python
# reading Outlier Removed data
df = pd.read_csv(r"G:\Uppsala Docs\period5&6\ProcessedData\2015-2019_sensorData.csv",
                 usecols=['Stop', 'Station Name','National Station Code','Longitude', 
                          'Latitude', 'Classification', 'NO2', 'NOX as NO2','PM2.5', 'CO',
                          'Black Carbon', 'O3',  'PM10'])

#2015-2019_OutliersRemoved_onSelectedStationAndFeatures.csv

# Rename Stop to Time
#df = df.rename(columns={"Timestamp":"DateTimestamp"})
df = df.rename(columns={"Stop":"DateTimestamp"})

# convert columns to date
df['DateTimestamp'] = pd.to_datetime(df['DateTimestamp'])

print(df.info())
print(df.describe())
```

```python
# merge additional features
df = df.merge(AirPressure ,on='DateTimestamp',how='outer')
df = df.merge(AirTemperature ,on='DateTimestamp',how='outer')
df = df.merge(CloudsAmount ,on='DateTimestamp',how='outer')
df = df.merge(PreciptationAmount ,on='DateTimestamp',how='outer')
df = df.merge(WindDirection ,on='DateTimestamp',how='outer')
# drop na in station column, remove 2020 data
df = df[~df[['Station Name']].isnull().any(1)]
print(df.info())
print(df.describe())
df.head(2)
```

```python
# fill na in additional columns using bfill
df[['Air pressure','Air temperature','Amount of clouds',
    'Precipitation amount','Wind direction','Wind speed']] = df[['Air pressure','Air temperature','Amount of clouds',
                                          'Precipitation amount','Wind direction','Wind speed']].fillna(method='bfill')
# df[['Air pressure','Air temperature','Amount of clouds',
#     'Precipitation amount','Wind direction','Wind speed']] = df[['Air pressure','Air temperature','Amount of clouds',
#                                           'Precipitation amount','Wind direction','Wind speed']].fillna(method='ffill')

print(df.info())
print(df.describe())
df.head(2)
```
